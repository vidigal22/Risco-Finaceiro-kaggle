1- Correlação entre Variáveis

```python
df = pd.read_csv('Dados_BR2.csv', encoding='latin1')

import pandas as pd

df_numerico = df.select_dtypes(include=[float, int])

# Calcular a correlação
correlacao = df_numerico.corr()
print(correlacao)
```

### Análise dos Valores de Correlação

1- Idade vs. Renda (0.002009)
- Correlação próxima de zero, indicando que não há uma relação linear significativa entre a idade e a renda.

2- Renda vs. Pontuação de Crédito (0.015511)
- Correlação muito baixa, indicando pouca ou nenhuma relação linear entre renda e pontuação de crédito.

3- Pontuação de Crédito vs. Montante do Empréstimo (-0.005996)
- Correlação próxima de zero, indicando pouca ou nenhuma relação linear entre a pontuação de crédito e o montante do empréstimo.

4- Montante do Empréstimo vs. Relação Dívida/Renda (0.026584)
- Correlação muito baixa, sugerindo que o montante do empréstimo não está linearmente relacionado à relação dívida/renda.

5- Anos no Emprego Atual vs. Relação Dívida/Renda (0.004973)
- Correlação próxima de zero, indicando pouca ou nenhuma relação linear entre os anos no emprego atual e a relação dívida/renda.

6- Valor dos Ativos vs. Número de Dependentes (-0.002776)
Correlação próxima de zero, indicando pouca ou nenhuma relação linear entre o valor dos ativos e o número de dependentes.

2- Visualização de Distribuições

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Distribuição de Idade
sns.histplot(df['Idade'], bins=100)
plt.title('Distribuição de Idade')
plt.show()

# Distribuição de Renda
sns.histplot(df['Renda'], bins=30)
plt.title('Distribuição de Renda')
plt.show()
```

3- Análise de Relação Dívida/Renda por Classificação de Risco

```python
sns.boxplot(x='Classificação de risco', y='Relação Divida/Renda', data=df)
plt.title('Relação Dívida/Renda por Classificação de Risco')
plt.show()
```

4- Impacto do Histórico de Pagamento na Classificação de Risco

```python
sns.boxplot(x='Classificação de risco', y='Histórico de Pagamento', data=df)
plt.title('Histórico de Pagamento por Classificação de Risco')
plt.show()
```
